# 데이터 전처리 학습

- pandas를 이용한 데이터 전처리 문법

- 1 query() : 행 추출
- 2 df[] : 열 추출
- 3 sort_values() : 정렬
- 4 groupby() : 집단 별로 나누기
- 5 assign() : 변수 추가
- 6 agg() : 통계치 구하기
- 7 merge() : 데이터 합치기(열)
- 8 concat() : 데이터 합치기(행)
