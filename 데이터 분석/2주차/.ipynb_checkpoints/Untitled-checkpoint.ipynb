## 데이터 전처리에 필요한 스킬

- pandas를 이용해서 데이터를 전처리하는데 필요한 문법
- query(), 행 추출
- df[], 열 추출
- sort_values() 정렬
- groupby() 집단 별로 나누기
- assign() 변수 추가
- agg() 통계치 구하기
- merge() 데이터 합치기(열)
- concat() 데이터 합치기(행)

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
```

```python
exam.query('nclass==1') # .을 이용해서 문법을 건다. .을 이용해서 문법을 확장할 수 있다
```

```python
exam.query('math==50')
```

```python
exam.query('nclass!=1') #1반을 제외한 나머지 반 모두 출력
```

```python
# 비교 연산자도 사용할 수 있음
# 이상, 이하도
exam.query('math>50')
```

```python
exam.query('math<=50')
```

```python
# 논리 연산을 이용해서 여러 가지 행 추출
exam.query('nclass==2 & math>=50') # 다양한 조건을 걸어서 데이터를 추출할 수 있다
```

```python
exam.query('nclass==2 | math>=80') # or 조건 
```

```python
exam.query('nclass==1 | nclass==2 | nclass==3') # or 조건을 이용한 다중 조건
```

### 문자열도 가능하다

```python
df = pd.DataFrame({'name':['A','B','C','D'],
                 'country' : ['K','J','E','F']})
```

```python
df.query('name=="A" & country=="K"') #문자열 데이터도 동일하게 적용 가능
```

```python
# 외부 변수를 이용한 추출도 가능
var_df=1
exam.query('nclass==@var_df') # 1반만 출력하라 "@"
```

# 순서대로 정렬하기
- 오름차순, 내림차순

```python
exam.sort_values('english') # 기본적으로 오름차순이 default
```

```python
exam.sort_values('english', ascending=False) # 내림차순
```

```python
# 여러 개의 컬럼을 정렬할 경우
exam.sort_values(['nclass','english']) #앞에 먼저 쓴 변수를 정렬 기준으로
```

- 오름차순과 내림차순을 마음대로 적용

```python
exam.sort_values(['nclass','english'], ascending=[True,False]) #변수에 따라 다르게 하고 싶은 경우
```

# numpy where() 이용해서 조건에 따른 값을 부여한 변수 만들기

```python
import numpy as np

exam.assign(test = np.where(exam['science'] >=50,'pass','fail'))
```

```python
np.where(exam['science']>=50) # 배열로 반환
```

```python
exam.assign(test= np.where(exam['science']>=50, 'pass', 'fail'))
```

```python
# lambda 식을 이용한 파생 변수 만들기
exam.assign(new = lambda x : x['math']+x['english']+x['science'])
```
